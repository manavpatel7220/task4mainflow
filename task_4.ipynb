# **TASK 4**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Create a simple sample dataframe
data = {
    'Age': [23, 45, 12, 36, 22, 85, 55, 29, 30, 40],
    'Salary': [40000, 50000, 12000, 36000, 22000, 85000, 55000, 29000, 30000, 40000],
    'Height': [5.1, 5.5, 4.8, 5.9, 5.3, 6.0, 5.8, 5.2, 5.4, 5.6]
}
```

```python
df = pd.DataFrame(data)
```

# Distribution of variables

```python
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
```

```python
plt.subplot(1, 3, 2)
sns.histplot(df['Salary'], kde=True)
plt.title('Distribution of Salary')
```

```python
plt.subplot(1, 3, 3)
sns.histplot(df['Height'], kde=True)
plt.title('Distribution of Height')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Boxplots to identify outliers
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 3, 1)
sns.boxplot(y=df['Age'])
plt.title('Boxplot of Age')
```

```python
plt.subplot(1, 3, 2)
sns.boxplot(y=df['Salary'])
plt.title('Boxplot of Salary')
```

```python
plt.subplot(1, 3, 3)
sns.boxplot(y=df['Height'])
plt.title('Boxplot of Height')
```

```python
plt.tight_layout()
plt.show()

```

```python
# Correlation matrix
corr_matrix = df.corr()
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```
